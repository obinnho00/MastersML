### a. Three people flip a fair coin. What is the probability that exactly two of them will get heads?

This follows the **binomial probability formula**:
\[
P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}
\]
Where:
- \( n = 3 \) (number of trials),
- \( k = 2 \) (number of successes),
- \( p = 0.5 \) (probability of heads).

\[
P(X = 2) = \binom{3}{2} (0.5)^2 (1-0.5)^{3-2}
\]
\[
P(X = 2) = 3 \times (0.5)^2 \times (0.5)^1 = 0.375
\]

**Answer:** \( P(X = 2) = 0.375 \) or **37.5%**.



### b. Twenty people flip a fair coin. What is the probability that exactly eight of them will get heads?

Using the same binomial probability formula:
\[
P(X = 8) = \binom{20}{8} (0.5)^8 (0.5)^{20-8}
\]
\[
P(X = 8) = \binom{20}{8} (0.5)^{20}
\]

The binomial coefficient is:
\[
\binom{20}{8} = \frac{20!}{8!(20-8)!} = 125,970
\]

\[
P(X = 8) = 125,970 \times (0.5)^{20} \approx 0.1201
\]

**Answer:** \( P(X = 8) \approx 0.1201 \) or **12.01%**.

### c. Twenty people flip a fair coin. What is the probability that at least four of them will get heads?

To calculate \( P(X \geq 4) \), use the complement rule:
\[
P(X \geq 4) = 1 - P(X \leq 3)
\]

Where:
\[
P(X \leq 3) = P(X = 0) + P(X = 1) + P(X = 2) + P(X = 3)
\]

Each term is calculated using the binomial formula:
\[
P(X = 0) = \binom{20}{0} (0.5)^0 (0.5)^{20} = (0.5)^{20}
\]
\[
P(X = 1) = \binom{20}{1} (0.5)^1 (0.5)^{19} = 20 \times (0.5)^{20}
\]

Compute \( P(X \leq 3) \), then subtract from 1. The result is:
\[
P(X \geq 4) \approx 0.9999
\]

**Answer:** \( P(X \geq 4) \approx 0.9999 \) or **99.99%**.

To model the descriptive features using a Naive Bayes model, we assume that each feature follows a **normal distribution**. The probability density function for a feature \( x \) given class \( C \) is:

\[
P(x \mid C) = \frac{1}{\sqrt{2 \pi \sigma_C^2}} \exp\left(-\frac{(x - \mu_C)^2}{2 \sigma_C^2}\right)
\]

Where:
- \( \mu_C \): Mean of the feature for class \( C \)
- \( \sigma_C^2 \): Variance of the feature for class \( C \)
