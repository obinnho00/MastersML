```python
from collections  import Counter # this is use to cound individual words 
import math 
import numpy as np
```

```python
# tokenize words 
count = Counter
words = ('o','x','y','m','o','r','o','n')
count_list = count(words) # Count the frequency oof each words 
total = len(words)
print(count_list)
print(total)
```

```python
entropy = -sum( (count/total) * math.log2(count/ total) for count in count_list.values())
print(entropy)
```

# 2

```python
vowels = "aeiouAEIOU"
Vowels = []
con = []
for word in words:
    if word in vowels:
        Vowels.append(word)
    elif word.isalpha():
        con.append(word)
v_count = count(Vowels)
con_count = count(con)
print(v_count)
print(con_count)
```

```python
Vowls_entropy = -sum( (val/len(Vowels)) * math.log2(val/len(Vowels))  for val in v_count.values())
print(Vowls_entropy )
```

```python
#  entropy for con 
con_entropy = -sum( (val/len(con)) * math.log2(val/len(con))  for val in con_count.values())
print(con_entropy )
```

```python
reduction_in_entropy = entropy - (Vowls_entropy  + con_entropy)
print(reduction_in_entropy)
```

# Using this dataset, construct the decision tree that would be generated by the ID3 algorithm, using entropy-based information gain.

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
import math 
```

```python
import warnings
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree

class Node:
    def __init__(self, feature=None, value=None, result=None, true_branch=None, false_branch=None):
        self.feature = feature
        self.value = value
        self.result = result
        self.true_branch = true_branch
        self.false_branch = false_branch
        self.dataset = None
        self.model = None

    def load_dataset(self, data):
        if data:
            self.dataset = pd.read_csv(data)
        else:
            print("Dataset not found")

    def split_and_train(self):
        if self.dataset is None:
            print("Dataset not loaded.")
            return

        X = self.dataset[['Good Behavior', 'Age Below 30', 'Drug Dependent']].values  # Use numpy array
        y = self.dataset['Recidivist'].values  # Use numpy array
        self.model = DecisionTreeClassifier(criterion='entropy')
        self.model.fit(X, y)

    def visualize_tree(self):
        if self.model is None:
            print("Model is not trained.")
            return

        plt.figure(figsize=(12, 8))
        plot_tree(self.model, feature_names=['Good Behavior', 'Age Below 30', 'Drug Dependent'], 
                  class_names=['False', 'True'], filled=True)
        plt.show()


class ModelPrediction:
    def __init__(self, model):
        self.model = model

    def crime_prediction(self, val1, val2, val3):
        new_data = [[val1, val2, val3]]  
        with warnings.catch_warnings():
            warnings.simplefilter("ignore")  
            prediction = self.model.predict(new_data)
        return prediction[0]


if __name__ == "__main__":
    node = Node()
    node.load_dataset("decision_tree_data.csv")

   
    node.split_and_train()
    node.visualize_tree()

    predictor = ModelPrediction(node.model)
    prediction1 = predictor.crime_prediction(False, False, True)
    prediction2 = predictor.crime_prediction(True, True, False)
    
```

# 2 What prediction will the decision tree generated in part (a) of this question return for the following query?
    Good Behavior = false, Age Below 30 = false, Drug Dependent = true

```python
print(f"Prediction for (Good Behavior=False, Age Below 30=False, Drug Dependent=True): {prediction1}")
```

```python
print(f"Prediction for (Good Behavior=True, Age Below 30=True, Drug Dependent=False): {prediction2}")
```
